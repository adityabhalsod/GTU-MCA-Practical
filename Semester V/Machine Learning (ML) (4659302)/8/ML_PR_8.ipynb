```python
#import library 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
#load dataset
df = pd.read_csv('dataset.csv')
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df['Interest Rate'], y=df['Median Home'], hue=df['Interest Rate'])
```

```python
sns.pairplot(data = df)
```

```python
sns.distplot(df['Interest Rate'])
```

```python
sns.distplot(df['Median Home'])
```

```python
x = df['Interest Rate'].values
y = df['Median Home'].values
x_data = np.array(x)
y_data = np.array(y)
x_data = x_data.reshape(-1,1)
y_data = y_data.reshape(-1,1)
```

```python
#import library for model split
from sklearn.model_selection import train_test_split
```

```python
#modal split
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.33, random_state=42)
```

```python
lm = LinearRegression()
```

```python
print(lm.fit(X_train,y_train))
```

```python
#final output
predict = lm.predict(X_test)
print(predict)
```

```python
# Plot outputs
plt.scatter(X_test, y_test,  color='blue')
plt.plot(X_test, lm.predict(X_test), color='red', linewidth=3)
plt.show()
```
